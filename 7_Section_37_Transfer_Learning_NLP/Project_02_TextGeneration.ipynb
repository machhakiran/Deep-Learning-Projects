```python
from transformers import pipeline

import textwrap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pprint import pprint
```

```python
text_gen = pipeline("text-generation")
```

```python
sentence = " Deep Learning has revolutionized Artificial Intelligence"  # you can put your large text as well.
```

```python
text_gen(sentence)
```

```python
print(text_gen(sentence, max_length=200))
```

```python
pprint(text_gen(sentence, max_length=200)) # pretty printer
```

```python
pprint(text_gen(sentence, num_return_sequences=3, max_length=40)) # Text generation is random because
                                                                   # it is generated from probability distribution
```

```python
sentence  = "The quick brown fox jumps over the lazy dog and the quick white cat jumps over the lazy rabbit"
```

```python
pprint(text_gen(sentence, num_return_sequences=5, max_length=40))
```
