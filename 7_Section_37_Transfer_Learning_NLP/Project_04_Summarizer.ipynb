```python
from transformers import pipeline
```

```python
string = 'Artificial Intelligence (AI) refers to the development of computer systems that can perform tasks that typically' +\
'require human intelligence. These tasks include learning, reasoning, problem-solving, perception, speech recognition, and'+\
'language understanding. The goal of AI is to create machines that can simulate or replicate human-like cognitive functions.' +\
 'Narrow or Weak AI is designed and trained for a specific task. It excels at performing a particular function but lacks' +\
 'the broad cognitive abilities of humans. Examples of narrow AI include virtual personal assistants, image and speech' +\
 'recognition systems, and recommendation algorithms. General or Strong AI would possess the ability to understand, learn' +\
 'and apply knowledge across a wide range of tasks, similar to human intelligence. Achieving true general AI remains a'+\
 'challenging and aspirational goal in the field.'
```

```python
summarizer = pipeline("summarization")
```

```python
summarizer(string)
```

```python
text = 'Image processing is a field of study within computer science and engineering that involves the manipulation of images' +\
'to extract information, enhance visual quality, or perform other tasks. It encompasses a wide range of techniques and' +\
'algorithms designed to analyze, alter, and interpret visual information. Image processing is used in various applications'+\
'including medical imaging, computer vision, remote sensing, and digital photography.'
```

```python
output = summarizer(text, min_length = 15, max_length = 35)[0]
```

```python
output['summary_text']
```
