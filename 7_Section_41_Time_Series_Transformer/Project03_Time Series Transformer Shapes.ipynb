```python
import torch
import torch.nn as nn
```

# Create the instance of Transformer Model

```python
transformer_model = nn.Transformer(d_model = 256, nhead = 8, num_encoder_layers = 6, num_decoder_layers = 6)
```

```python
transformer_model
```

# Create the source and target value for the tranformer

```python
src = torch.rand((200, 256))
tgt = torch.rand((1, 256))
```

```python
 out = transformer_model(src, tgt)
```

```python
out.shape   # equal to the shape of the target
```
